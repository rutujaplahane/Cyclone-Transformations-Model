```python
import os
import copernicusmarine
from datetime import datetime

# --- 1. Define Download Parameters ---

DATASET_ID = "cmems_obs-mob_glo_phy_my_0.125deg_P1D-m"
OUTPUT_DIR = r"E:\ML Project\WOSC Project\EarthFormer\Data"  # Directory to save the NetCDF files

# Geographical and Depth Constraints
MIN_LON = 82.5
MAX_LON = 90.0
MIN_LAT = 15.0
MAX_LAT = 22.5
MIN_DEPTH = 0.0
MAX_DEPTH = 700.0

# Variables to download (e.g., Temperature, Salinity)
# You should check the dataset documentation for the exact variable names
# For "phy" (physical) data, common variables are 'thetao' (temperature) and 'so' (salinity).
# I'll use placeholders here - replace these with the actual variable names you need!
VARIABLES = ["to", "so", "ugo", "vgo", "zo"]

# Time Periods (Replace these with your actual 6 time periods)
# Format: [ (start_datetime, end_datetime), ... ]
# Datetimes should be strings in ISO format: "YYYY-MM-DDTHH:MM:SS" or "YYYY-MM-DD" for daily data
TIME_PERIODS = [
    ("Phailin", "2013-09-24", "2013-10-23"),
    ("Hudhud",  "2014-09-27", "2014-10-26"),
    ("Asani",   "2022-04-27", "2022-05-26"),
    ("Titli",   "2018-09-28", "2018-10-27"),
    ("Fani",    "2019-04-16", "2019-05-15"),
    ("Amphan",  "2020-05-06", "2020-06-04"),
    ("Yaas",    "2021-05-13", "2021-06-11"),
    ("Dana",    "2024-10-12", "2024-11-10"),
    ("Gulab",   "2021-09-14", "2021-10-13"),
    ("Montha",  "2025-10-15", "2025-11-13"),
]

# --- 2. Create Output Directory ---
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"Output directory '{OUTPUT_DIR}' ensured.")

# --- 3. Download Loop ---
print(f"Starting download of data set: {DATASET_ID}...")

for i, (cyclone_name, start_dt, end_dt) in enumerate(TIME_PERIODS, 1):
    # Construct a unique output filename for each period
    
    safe_name = cyclone_name.lower()
    filename = f"{safe_name}_{start_dt}_to_{end_dt}.nc"
    output_path = os.path.join(OUTPUT_DIR, filename)

    print("-" * 50)
    print(f"Downloading {cyclone_name} ({start_dt} → {end_dt})")
    print(f"Saving to: {output_path}")


    try:
        copernicusmarine.subset(
            dataset_id=DATASET_ID,
            variables=VARIABLES,
            minimum_longitude=MIN_LON,
            maximum_longitude=MAX_LON,
            minimum_latitude=MIN_LAT,
            maximum_latitude=MAX_LAT,
            minimum_depth=MIN_DEPTH,
            maximum_depth=MAX_DEPTH,
            start_datetime=start_dt,
            end_datetime=end_dt,
            output_filename=output_path,
            force_download=True          # Forces download without asking for confirmation
        )
        print(f"Successfully downloaded Period {i}.")
       
    except Exception as e:
        print(f"⚠️ Error downloading Period {i}: {e}")
        # Continue to the next period even if one fails
        continue

print("=" * 50)
print("All download requests completed.")
```
